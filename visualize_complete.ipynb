```python
import plotly.express as px
import pandas as pd
import dask.dataframe as dd
```

```python
df = dd.read_csv('data/*.csv').compute()
df
query = "continent=='Africa'"
report_name = "Hello"
```

```python
print(report_name)
```

```python
df = df.query(query)
```

```python

fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="country",
           hover_name="country", log_x=True, size_max=90, animation_frame="year")
fig.show()
```

```python
fig = px.sunburst(df.query('year == 2007'), path=['continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
```

```python
fig = px.treemap(df.query('year == 2007'), path=[px.Constant('world'), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
```

```python
fig = px.choropleth(df, locations="iso_alpha", color="lifeExp", hover_name="country", animation_frame="year", range_color=[20,80])
fig.show()
```
