```python
import pandas as pd
import re
import plotly.express as px
from IPython.display import display
```

```python
df = px.data.gapminder()
```

```python
for cnty in df.country.unique():
    cnty_clean = re.sub('\W','',cnty)
    print(cnty_clean)
    df[df['country'] == cnty].to_csv(f"data/{cnty_clean}.csv",index=False)
```
